```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset (update the filename if needed)
file_path = "../data/raw/smartwatch_sleep.csv"  # or change to match your filename
df = pd.read_csv("/Users/rutujashingate/Desktop/sleep-ml-project/data/raw/Sleep_health_and_lifestyle_dataset.csv")

# Show the first 5 rows
df.head()
```

```python
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df.drop('Person ID', axis=1, inplace=True)
```

```python
# Fill NaNs in 'Sleep Disorder' with 'None'
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')
```

```python
# Split into two numeric columns
df[['Systolic_BP', 'Diastolic_BP']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)

# Drop the original
df.drop('Blood Pressure', axis=1, inplace=True)
```
